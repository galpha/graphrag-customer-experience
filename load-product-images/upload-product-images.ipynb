# Copy Article Images to UI Folder and Serve

```python
from langchain_community.graphs import Neo4jGraph
from dotenv import load_dotenv
import os

load_dotenv('.env', override=True)

# Neo4j
NEO4J_URI = os.getenv('NEO4J_URI')
NEO4J_USERNAME = os.getenv('NEO4J_USERNAME')
NEO4J_PASSWORD = os.getenv('NEO4J_PASSWORD')

#OPENAI
OPENAI_API_KEY = os.getenv('OPENAI_API_KEY')

graph = Neo4jGraph()
```

```python
article_ids = graph.query('MATCH(a:Article) RETURN collect(a.articleId) AS articleIds')[0]['articleIds']
len(article_ids)
```

```python
article_ids[:10]
```

```python
a = article_ids[0]

f'h-and-m-personalized-fashion-recommendations/images/0{str(a)[:2]}/{a}.jpg'
```

```python
# import module
import shutil

# copy the contents of the demo.py file to  a new file called demo1.py
failed_article_ids = []
for a in article_ids:
    try:
        src = f'h-and-m-personalized-fashion-recommendations/images/0{str(a)[:2]}/0{a}.jpg'
        tgt = f'ui/images/{a}.jpg'
        shutil.copyfile(src, tgt)
    except Exception as e:
        failed_article_ids.append(a)
        src = 'photo-coming-soon.jpg'
        tgt = f'ui/images/{a}.jpg'
        shutil.copyfile(src, tgt)
        print(e)
```
